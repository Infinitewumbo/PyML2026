### Objective

```python
# Solution
def reflect_on_draft(draft: str, model: str = "openai:o4-mini") -> str:

    ### START CODE HERE ###

    prompt = f"""
You are a writing coach. Read the essay draft below and provide critical but constructive feedback in paragraph form.

Focus on:
- Structure: introduction, thesis clarity, body organization, conclusion
- Clarity: unclear sentences, confusing phrasing, missing context
- Strength of argument: reasoning, evidence/examples, counterpoints, logical flow
- Writing style: tone, transitions, repetition, grammar/mechanics

Do NOT rewrite the essay. Do NOT provide a revised version. Only give feedback.

Essay draft: {draft}
"""

    ### END CODE HERE ###

    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    return response.choices[0].message.content
```

```python
# Solution
def revise_draft(original_draft: str, reflection: str, model: str = "openai:gpt-4o") -> str:

    ### START CODE HERE ###

    prompt = f"""
        You are an expert writing assistant. Revise the essay draft below using the feedback provided.

        Goals:
        - Address the specific issues mentioned in the feedback
        - Improve clarity, coherence, and flow
        - Strengthen the argument (better reasoning, examples, and organization)
        - Improve transitions and writing style
        - Keep the same topic and overall intent, but rewrite as needed

        Return ONLY the revised essay (no bullet points, no explanations, no commentary).

        Feedback: {reflection}

        Original draft:{original_draft}
        """

    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    ### END CODE HERE ###

    return response.choices[0].message.content
```
